```python
import matplotlib.pyplot as plt
import numpy as np
```

# Problem Set 1: Truncation Error of Forward and Central Differences

```python
forward_data = np.genfromtxt('forward_error.txt', delimiter=' ')
central_data = np.genfromtxt('central_error.txt', delimiter=' ')
```

```python
stepsizes = forward_data[:, 0]
forward_error = forward_data[:, 1]
central_error = central_data[:, 1]
log_forward_error = np.log(forward_error)
log_central_error = np.log(central_error)
log_stepsizes = np.log(stepsizes)
```

```python
plt.title('Finite Differences Error vs. Stepsize')
plt.plot(log_stepsizes, log_forward_error, label='Forward Difference', marker='o')
plt.plot(log_stepsizes, log_central_error, label='Central Difference', marker='o')
plt.xlabel('Log Stepsize')
plt.ylabel('Log Error')
plt.legend();
```

```python
forward_slope = (log_forward_error[3] - log_forward_error[0])/(stepsizes[3] - stepsizes[0])
forward_slope
```

```python
central_slope = (log_central_error[3] - log_central_error[0])/(stepsizes[3] - stepsizes[0])
central_slope
```
